```python
def splitDataFrameIntoSmaller(df, chunkSize): 
    listOfDf = list()
    numberChunks = len(df) // chunkSize + 1
    for i in range(numberChunks):
        listOfDf.append(df[i*chunkSize:(i+1)*chunkSize])
    return listOfDf
```

```python
split_df=splitDataFrameIntoSmaller(dept_age, 10)
```
